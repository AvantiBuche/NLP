```python
#stemming
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
sentences = nltk.sent_tokenize(paragraph)
```

```python
sentences
```

```python
len(sentences)
```

```python
for i in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i]) # [Thank, you, all, so, very, much]
    newwords = [stemmer.stem(word) for word in words] #list which contain stemm word
    sentences[i] = ' '.join(newwords)
```

```python
sentences
#the output is not that accurate - very, this
```

```python
# Lemmatizer
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
sentences = nltk.sent_tokenize(paragraph)
```

```python
sentences
```

```python
for i in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i])
    newwords = [lemmatizer.lemmatize(word) for word in words]
    sentences[i] = ' '.join(newwords)
```

```python
sentences
#not all the sentences are included - was, parents
```

```python
# Stopwords - removes stop words
from nltk.corpus import stopwords
sentences = nltk.sent_tokenize(paragraph)
sentences
```

```python
stopwords.words('english')
```

```python
for i in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i]) #[Thank ,you ,all ,so ,very, much]
    newwords = [word for word in words if word not in stopwords.words('english')]
    sentences[i] = ' '.join(newwords)
```
